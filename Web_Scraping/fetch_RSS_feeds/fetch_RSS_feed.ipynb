```python
#import feedparser library
import feedparser
```

```python
#url for the RSS feed
url = "https://rss.app/feeds/EtK9r1UhbzUOsF24.xml"
```

```python
#methods and attributes
dir(feedparser)
```

```python
#parse the url 
f = feedparser.parse(url)
```

```python
f
```

```python
#Check the keys of feed
f.keys()
```

```python
#request header informations 
f.headers
```

```python
#news feed informations 
f.feed
```

```python
#news informations 
f.entries
```

```python
#no of news 
len(f.entries)
```

```python
#get titles of each news 
for entry in f.entries:
    print(entry.title)
```

```python
#Get first link for the real news 
data = f2.entries[0].links[0].href
print(data)
```

```python
#import libraries for web scraping to get details of news 
import nltk
from newspaper import Article
```

```python
url3 = data
```

```python
article = Article(url3)
```

```python
article.download()

nltk.download('punkt')
article.parse()
article.nlp()
```

```python
dir(article)
```

```python
#detailed information of news 
print(article.text)
```
